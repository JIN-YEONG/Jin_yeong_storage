```python
#-*- coding: utf-8 -*-


# 주식 데이터를 이용해 다음날 종가 예측

import pandas as pd

from sklearn.preprocessing import MinMaxScaler

file_data = pd.read_csv('D:/study/keras/0801/data/kospi200test.csv', encoding='euc-kr')

print(file_data.head())


```

```python
file_data.head()
```

```python
file_data.drop('Unnamed: 7' ,axis=1, inplace=True)

file_data.head()
```

```python
file_data=file_data.sort_values(by=['일자']) # 시간순으로 정렬
```

```python
file_data.head()
```

```python
file_data.drop('일자', axis=1, inplace=True)
file_data
```

```python
price = file_data.values[:,:-1]   # 환율을 빼고 정규화
scaler = MinMaxScaler()
scaler.fit(price)
mm_price = scaler.transform(price)
mm_price
```

```python
volume = file_data.values[:,-1:]   # 환율 정규화
scaler.fit(volume)
mm_volume = scaler.transform(volume)
mm_volume
```

```python
import numpy as np
x = np.concatenate((mm_price, mm_volume), axis=1)
x
```

```python
y = x[:,3]
y
```

```python
data_x = []
data_y = []

for i in range(0, len(y) - 5):
    data_x.append(x[i:i+5])
    data_y.append(y[i+5])
    
data_x = np.array(data_x)
data_y = np.array(data_y)

print(data_x.shape)
data_y.shape
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x , train_y, test_y = train_test_split(
    data_x, data_y, random_state = 81, test_size=0.4
)

val_x, test_x ,val_y, test_y = train_test_split(
    test_x, test_y, random_state = 81, test_size = 0.5
)

print(train_x.shape)
print(train_y.shape)
print('-'*10)
print(test_x.shape)
print(test_y.shape)
print('-'*10)
print(val_x.shape)
print(val_y.shape)
```
